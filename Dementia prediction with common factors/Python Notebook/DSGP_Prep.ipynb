```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('dataset.csv')

# Check for null values
null_values = df.isnull().sum()

# Display columns with null values
null_columns = null_values[null_values > 0]
if null_columns.empty:
    print("No null values found in the dataset.")
else:
    print("Columns with null values:")
    print(null_columns)
```

```python
df.dropna(inplace=True)

# Verify the changes
print(df.isnull().sum())
```

```python
# prompt: I want to check duplicate rows

# Check for duplicate rows
duplicate_rows = df[df.duplicated()]

if duplicate_rows.empty:
    print("No duplicate rows found in the dataset.")
else:
    print("Duplicate rows:")
duplicate_rows
```

```python
len(duplicate_rows)
```

```python
# Drop duplicate rows
df = df.drop_duplicates()

# Check for duplicate rows again (optional, to verify)
duplicate_rows = df[df.duplicated()]

if duplicate_rows.empty:
    print("No duplicate rows found in the dataset after dropping duplicates.")
else:
    print("Duplicate rows (after dropping):") # This should not print anything if duplicates were removed successfully
    print(duplicate_rows)
    print(len(duplicate_rows))
```

```python
print("\nUnique value count in each column:")
print(df.nunique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Get continuous columns
continuous_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Iterate over continuous columns
for col in continuous_cols:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts(), "\n")  # Print count of unique values

    # Plot histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, bins=10)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Define the columns to replace values in
columns_to_replace = ['TOBAC100', 'Depression', 'NACCFAM', 'DIABETES', "HYPERCHO", 'HYPERTEN', 'ALCFREQ', 'CVHATT', 'CVCHF', 'MARISTAT']

for col in columns_to_replace:
    if col in df.columns:  # Check if the column exists in the DataFrame
        df[col] = df[col].replace([-4, 9], -1)
    else:
        print(f"Warning: Column '{col}' not found in the DataFrame.")
```
